```python
import pandas as pd
import numpy as np
```

```python
mall = pd.read_csv("Mall_Customers.csv")
mall
```

```python
mall.isna().sum()
```

```python
mall.describe()
```

```python
#Another clean dataset, nice to have
#Let's do some data visualisation

import matplotlib.pyplot as plt

plt.scatter(mall["Annual Income (k$)"], mall["Spending Score (1-100)"])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Visualization of data')
plt.show()
```

```python
plt.scatter(mall["Age"], mall["Spending Score (1-100)"])
plt.xlabel('Age')
plt.ylabel('Spending Score')
plt.title('Visualization of data')
plt.show()
```

```python
#We cant use gender for visualisation, because it is binary, and we cant use Customer ID because it doesn't form any
#relation with spending score

mall.drop(['CustomerID', 'Gender'], axis = 1, inplace = True)
mall
```

```python
#We will make 2 models for 2 different set of values, i.e., firstly we will generalise customers on basis of their relation 
#between annual income and spending score and then we will generalise them according to age and spending score
x1 = mall.filter(['Annual Income (k$)', 'Spending Score (1-100)']).values
x2 = mall.filter(['Age','Spending Score (1-100)']).values
x1
```

```python
x2
```

```python
#Now we need to determine k, for kmeans classification

from sklearn.cluster import KMeans, SpectralClustering
finding_elbow = []

#This function will help us to find the right k for our function
for i in range(1, 7):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 69)  
    kmeans.fit(x1)  
    finding_elbow.append(kmeans.inertia_)  
    
plt.plot(range(1,7), finding_elbow)
plt.xlabel('Number of clusters')
plt.ylabel('finding_elbow')
plt.show()
```

```python
#We can see the elbow is formed at 5


#Now lets make k means algorithm for 5 neighbours
kmeans = KMeans(n_clusters=5, init='k-means++', random_state= 69)  
y_pred = kmeans.fit_predict(x1)
```

```python
# for visulaizing the clusters, I always use this type of function, which I have seen from a website online
#Because it simplifies the graph drawing
plt.scatter(x1[y_pred == 0, 0], x1[y_pred == 0, 1], s = 100, c = 'red')
plt.scatter(x1[y_pred == 1, 0], x1[y_pred == 1, 1], s = 100, c = 'blue') 
plt.scatter(x1[y_pred == 2, 0], x1[y_pred == 2, 1], s = 100, c = 'black')
plt.scatter(x1[y_pred == 3, 0], x1[y_pred == 3, 1], s = 100, c = 'green') 
plt.scatter(x1[y_pred == 4, 0], x1[y_pred == 4, 1], s = 100, c = 'brown') 
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')  

plt.title("Segmentation according to Annual Income")
plt.show()
```

```python
#Now for x2
#This function will help us to find the right k for our function
finding_elbow = []
for i in range(1, 7):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 69)  
    kmeans.fit(x2)  
    finding_elbow.append(kmeans.inertia_)  
    
plt.plot(range(1,7), finding_elbow)
plt.xlabel('Number of clusters')
plt.ylabel('finding_elbow')
plt.show()
```

```python
#Its elbow is at 2

#Now lets make k means algorithm for 2 neighbours
kmeans = KMeans(n_clusters=2, init='k-means++', random_state= 69)  
y_pred = kmeans.fit_predict(x2)
```

```python
# for visulaizing the clusters, I always use this type of function, which I have seen from a website online
#Because it simplifies the graph drawing
plt.scatter(x2[y_pred == 0, 0], x2[y_pred == 0, 1], s = 100, c = 'red')
plt.scatter(x2[y_pred == 1, 0], x2[y_pred == 1, 1], s = 100, c = 'blue') 
 
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')  

plt.title("Segmentation according to Age")
plt.show()
```
